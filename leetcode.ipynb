````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        # Time Complexity O(n), Space Complexity O(1)
        if len(nums) == 1:
            return nums[0]

        dp0 = nums[0]
        dp1 = max(nums[0], nums[1])

        for i in range(2, len(nums)):
            ans = max(dp1, dp0+nums[i])
            dp0 = dp1
            dp1 = ans

        return dp1
    
#===================================================
        if len(nums) == 1:
            return nums[0]

        dp = [0] * len(nums)

        dp[0] = nums[0]
        dp[1] = max(dp[0], nums[1])

        for i in range(2, len(nums)):
            dp[i] = max(dp[i-1], dp[i-2]+nums[i])

        return dp[-1]

#=============================================
        #Time Complexity O(n), Space Complexity O(n)
        dp = {}
        def f(i):
            if i in dp:
                return dp[i]

            if i >= len(nums):
                return 0

    
            steal = f(i+2)+nums[i]
            nosteal = f(i+1)
            dp[i] = max(steal, nosteal)
            return max(steal, nosteal)

        return f(0)
    
    #========================================
        cache = [-1] * len(nums)
        def f(n):
            if n < 0:
                return 0
            if cache[n] != -1:
                return cache[n]

            choose_n = f(n-2) + nums[n]
            no_choose_n = f(n-1)
            cache[n] = max(choose_n, no_choose_n)
            return max(choose_n, no_choose_n)

        return f(len(nums) - 1)
````

````python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:

        dp = [float("INF")] * (amount + 1)

        dp[0] = 0

        for coin in coins:
            for i in range(coin, amount+1):
                dp[i] = min(dp[i], dp[i-coin]+1)

        return dp[amount] if dp[amount] != float("INF") else -1
        
        #=========================================
        #从大到小找
        dp = {}

        def f(i, amount):
            if i < 0:
                return 0 if amount == 0 else float("INF")

            if (i, amount) in dp:
                return dp[(i, amount)]

            if coins[i] > amount: #大于当前金额，不能选, 如果是从小到大找，则不是这个判断
                return f(i-1, amount) # return也不一样

            select_this = f(i, amount - coins[i]) + 1
            no_select_this = f(i-1, amount)
            
            dp[(i, amount)] = min(select_this, no_select_this)
            return dp[(i, amount)]

        ans = f(len(coins) - 1, amount)
        return ans if ans < float("INF") else -1

        #========================================
        
        #从小到大找
        dp = {}
        def f(i, amount):
            if (i, amount) in dp:
                return dp[(i, amount)]
            if i == len(coins):
                if amount == 0:
                    return 0
                else:
                    return float("INF")

            if amount < 0: # amount < 0, 不能选，如果是从大到小找，则不是这个判断
                return float("INF") # return也不一样

            select_i = f(i, amount - coins[i]) + 1
            no_select = f(i+1, amount)

            dp[(i, amount)] = min(select_i, no_select)
            return dp[(i, amount)]

        ans = f(0, amount)
        return ans if ans <float("inf") else -1
````

````python
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        
        ans = []
        path = []

        def f():
            if len(path) == len(nums):
                ans.append(path.copy())
                return

            for j in nums:
                if j not in path:
                    path.append(j)
                    f()
                    path.pop()

        f()
        return ans
    #=========================================
        #使用cache来判断比下面的if not in path更快，为O(1)，但总体时间复杂度还是O(n*n!)
        cache = [False] * len(nums)
        
        ans = []
        path = []

        def f(i):
            if i == len(nums):
                ans.append(path.copy())
                return

            for j in range(len(nums)):
                if not cache[j]:
                    path.append(nums[j])
                    cache[j] = True
                    f(i+1)
                    path.pop()
                    cache[j] = False
        f(0)
        return ans
    
        #=========================================
        # 递归时间复杂度为O(n!)，not in path 需要O(n), 所以最终时间复杂度为O(n*n!), 空间复杂度为O(n)
        ans = []
        path = []

        def f(i):
            if i == len(nums):
                ans.append(path.copy())
                return

            for num in nums:
                if num not in path:
                    path.append(num)
                    f(i+1)
                    path.pop()
        f(0)
        return ans
    #========================================
    # 空间时间复杂度同上，但是用myPath查询为O(1), 最终时间复杂度为O(n*n!), 空间复杂度为O(n)
        ans = []
        path = []

        myPath = [0] * len(nums)

        def f(i):
            if i == len(nums):
                ans.append(path[:])
                return

            for j in range(len(nums)):
                if not myPath[j]: # not used
                    path.append(nums[j])
                    myPath[j] = 1
                    f(i+1)
                    path.pop()
                    myPath[j] = 0

        f(0)
        return ans
````

````python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        # Divide array into two parts using mid
        # One part must be sorted, the other part then is rotated (Unless the original array is not rotated)
        # If the sorted part contains target so normal binary search

        # Else target must be in the other half (which may be rotated or sorted)
        #       We continue dividing, and the rotated part will again split into
        #       one sorted and one rotated part
        
        
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid
            if nums[left] <= nums[mid]:

                if nums[left] <= target <= nums[mid]: # This does not use slice so no extra space
                    right = mid - 1
                else:
                    left = mid + 1

            else:
                if nums[mid] <= target <= nums[right]:
                    left = mid + 1
                else:
                    right = mid - 1
        return -1
    
        # ==========================================

        def bs(nums, target):
            left = 0
            right = len(nums) - 1

            while left <= right:
                mid = (right - left) // 2 + left
                if nums[mid] < target:
                    left = mid + 1
                elif nums[mid] > target:
                    right = mid - 1
                else:
                    return mid
            return -1

        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (right - left) // 2 + left
            if nums[left] <= nums[mid]: # Left part is sorted
                if nums[left] <= target <= nums[mid]: # Target is in left part
                    idx = bs(nums[left:mid+1], target) # This uses slice, extra space O(n)
                    return left + idx if idx != -1 else -1

                else: # Target is in right part, but right part is not sorted
                    left = mid + 1

            else: # Right part is sorted
                if nums[mid] <= target <= nums[right]: # target is in right part
                    idx = bs(nums[mid:right+1], target)
                    return mid + idx if idx != -1 else -1
                else:
                    right = mid - 1

        return -1
````

````python
'''
    假设nums1的长度是m, nums2的长度是n, 
    当m+n为奇数, 中位数是第(m+n)/2的数
    当m+n为偶数, 中位数是第(m+n)/2和第(m+n)/2+1的平均值
    
    所以当 m+n 为奇数时，中位数是合并后数组的第 ⌊(m+n)/2⌋ 个元素（从 0 开始计数
    当 m+n 为偶数时，中位数是合并后数组的第 ⌊(m+n)/2⌋ 和第 ⌊(m+n)/2⌋ + 1 个元素的平均值
    
    定义 i 为 nums1 左部分的元素数量, nums1[0...i-1] 为左部分, nums1[i...m-1] 为右部分
    定义 j 为 nums2 左部分的元素数量, nums2[0...j-1] 为左部分, nums2[j...n-1] 为右部分
    
    因此两个条件找到中位数:
    1.
    奇数长度时，左部分元素数量为 (m+n+1)/2, 右部分为 (m+n-1)/2
    偶数长度时，左右部分各有 (m+n)/2 个元素
    左边长度又可以用i+j表示, 则i+j = (m+n+1) // 2, 中位数为左边最大 (偶数也用这个, 计算的时候取左边最大右边最小的平均)
    则j = (m+n+1) // 2 - i
    而i可以用二分查找找到
    
    2.
    nums1[i-1] <= nums2[j](nums1 左部分最大值小于等于 nums2 右部分最小值)
    nums2[j-1] <= nums1[i](nums2 左部分最大值小于等于 nums1 右部分最小值)
    
    比如:
    奇数时
    nums1 = [1, 3, 5, 7, 9] m = 5
    nums2 = [2, 4, 6, 8] n = 4
    m+n = 9
    左部分元素数量：(m+n+1)/2 = 5, 右部分: (m+n-1)/2 = 4

    迭代i的值(二分查找), j = (m+n+1)//2 - i, 直到满足条件1和条件2
    则i=3, j=2, i+j = 3+2 = 5, (m-i)+(n-j) = (5-3)+(4-2) = 4
    nums1 左部分: [1, 3, 5]  ，右部分: [7, 9]
    nums2 左部分: [2, 4]     ，右部分: [6, 8]
    中位数: 左部分最大值 max(5, 4) = 5
    
    偶数时
    nums1 = [1, 3, 5, 7] m = 4
    nums2 = [2, 4, 6, 8] n = 4
    m+n = 8
    左部分元素数量：(m+n)/2 = 4, 右部分: (m+n)/2 = 4

    则i=2, j=2, i+j = 2+2 = 4, (m-i)+(n-j) = (5-2)+(4-2) = 4
    nums1 左部分: [1, 3]     ，右部分: [5, 7]
    nums2 左部分: [2, 4]     ，右部分: [6, 8]
    中位数: (左部分最大值 + 右部分最小值) / 2 = (max(3, 4) + min(5, 6)) / 2 = (4 + 5) / 2 = 4.5
    
    边界条件:
    i == n, j == m, 值为inf
    i == 0, j == 0, 值为-inf

    
'''
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        #确保在最短上找，减少时间复杂度，并且如果nums1比nums2长，i的值可能超过nums1的长度
        if len(nums1) > len(nums2):
            return self.findMedianSortedArrays(nums2, nums1)

        m, n = len(nums1), len(nums2)

        left = 0
        right = m

        while left <= right:
            i = (left+right) // 2
            j = (m+n+1)//2 - i

            nums1_left = -float("INF") if i == 0 else nums1[i-1]
            nums2_left = -float("INF") if j == 0 else nums2[j-1]

            nums1_right = float("INF") if i == m else nums1[i]
            nums2_right = float("INF") if j == n else nums2[j]

            if nums1_left <= nums2_right and nums2_left <= nums1_right:
                if not (m+n) % 2: #even
                    return float((max(nums1_left, nums2_left) + min(nums1_right, nums2_right)) / 2)

                else: #odd
                    return max(nums1_left, nums2_left)

            elif nums1_left > nums2_right:
                right = i - 1
            else:
                left = i + 1
````

````python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        
        f0 = 0
        f1 = 0

        for i in range(2, len(cost)+1):
            take1step = f1 + cost[i-1]
            take2step = f0 + cost[i-2]
            ans = min(take1step, take2step)
            f0 = f1
            f1 = ans

        return f1
    
#==========================================
        # Why len(cost)+1? Because we need to reach the step after the last step
        f = [0] * (len(cost)+1)

        for i in range(2, len(cost)+1):
            take1step = f[i-1] + cost[i-1]
            take2step = f[i-2] + cost[i-2]
            f[i] = min(take1step, take2step)

        return f[-1]

#==========================================
        dp = {}
        def f(i):
            if i in dp:
                return dp[i]

            # Why i <= 1 not i < 0? Because we can start from step 0 or step 1 directly, that means, we can reach step 0 or step 1 with 0 cost according to the problem description.
            if i <= 1:
                return 0

            select1 = f(i-1) + cost[i-1]
            select2 = f(i-2) + cost[i-2]

            dp[i] = min(select1, select2)
            return dp[i]
        return f(len(cost))
````

````python
class DListNode:
    def __init__(self, key, val):
        self.key = key
        self.val  = val
        self.next = None
        self.prev = None


class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.dic = {}
        self.head = DListNode(-1, -1)
        self.tail = DListNode(-1, -1)
        self.head.next = self.tail
        self.tail.prev = self.head
        

    def get(self, key: int) -> int:
        if key not in self.dic:
            return -1

        node = self.dic[key]
        self.remove(node)
        self.add(node)
        return node.val

    def put(self, key: int, value: int) -> None:
        if key in self.dic:
            old_node = self.dic[key]
            self.remove(old_node)

        node = DListNode(key, value)
        self.dic[key] = node
        self.add(node)

        if len(self.dic) > self.capacity:
            node_to_delete = self.head.next
            self.remove(node_to_delete)
            del self.dic[node_to_delete.key]

    def add(self, node):
        previous_end = self.tail.prev
        previous_end.next = node
        node.prev = previous_end
        node.next = self.tail
        self.tail.prev = node
        
    def remove(self, node):
        node.prev.next = node.next
        node.next.prev = node.prev



# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
````

````python
class Solution:
    def longestPalindrome(self, s: str) -> str:

        ans_left = 0
        ans_right = 0

        # Odd situation
        for i in range(len(s)):
            l = i
            r = i

            while l >= 0 and r < len(s) and s[l] == s[r]:
                # When this stop, l and r is beyond the palindromic index, so actual index for palindromic is l+1 and r-1
                l-=1
                r+=1

            # No more palindromic, end and compare length of subtring
            if r-1-(l+1)+1 > ans_right-ans_left+1:
                ans_right = r-1
                ans_left = l+1

        # Even situation
        for i in range(len(s)):
            l = i
            r = i+1

            while l >= 0 and r < len(s) and s[l] == s[r]:
                # When this stop, l and r is beyond the palindromic index, so actual index for palindromic is l+1 and r-1
                l-=1
                r+=1

            # No more palindromic, end and compare length of subtring
            if r-1-(l+1)+1 > ans_right-ans_left+1:
                ans_right = r-1
                ans_left = l+1

        return s[ans_left:ans_right+1]
````

````python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        k %= len(nums) # Avoid case like [-1], k = 2

        # Do not use [::-1], because it will create a new list
        # Do not use nums[k:] + nums[:k], because it will create a new list
        # Do not use reversed(), because it will create a new iterator
        def reverse(left, right):
            while left < right:
                nums[left], nums[right] = nums[right], nums[left]
                left += 1
                right -= 1

        reverse(0, len(nums) - 1)
        reverse(0, k-1)
        reverse(k, len(nums)-1)
````

````python
class Solution:
    def jump(self, nums: List[int]) -> int:
        # We are gonna maintain two nxt, the current max reachable index, and the next max reachable index (Global Max)
        mx_nxt = 0 # This is the global max reachable (This global means from index 0 to i, what is the max reachable index, not 0 to end)
        cur_nxt = 0 # This is the current max reachable (This current means from the last jump, what is the max reachable index)
        ans = 0

        for i in range(len(nums)-1):
            mx_nxt = max(mx_nxt, i+nums[i])
            if i == cur_nxt: # If i equals to where we already reach, that means we reach the boundary, and we cannot move forward anymore, so we need to jump, but jump from where and to where?
                cur_nxt = mx_nxt # This line means: I cannot move forward anymore, so I GO BACK to where I can jump to the global max reachable, and I jump to it, so ans += 1
                ans += 1
        return ans
    
    # Here is some possible confusion:
    # 1. Why for i in range(len(nums)-1): not for i in range(len(nums)):
    # Because when we are at the last index, we do not need to jump anymore, so we do not need to consider the last index.
    
    # 2. Why greedy can find the optimal solution?
    # Because we are not greedy about single index, but all indices from last jump to current index, we are greedy about the global max reachable from all these indices, so we can always find the optimal solution.
````

````python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:

        def bs(nums: List[int], target: int) -> int:

            left = 0
            right = len(nums) - 1

            while left <= right:
                mid = left + (right - left) // 2
                if nums[mid] > target:
                    right = mid - 1
                elif nums[mid] < target:
                    left = mid + 1
                else:
                    return True
            return False

        left = 0
        right = len(matrix) - 1

        while left <= right:
            mid = (right-left) // 2 + left

            if matrix[mid][0] <= target and matrix[mid][-1] >= target:
                return bs(matrix[mid], target)
            elif matrix[mid][0] > target:
                right = mid - 1
            else:
                left = mid + 1
        return False
````

````python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        rows = [set() for _ in range(9)] 
        # Do not use [set() * 9] that means the element in set multiply by 9
        cols = [set() for _ in range(9)]
        subbox = [[set() for _ in range(3)]for _ in range(3)] 

        for r in range(len(board)):
            for c in range(len(board[0])):
                val = board[r][c]
                if val == '.':
                    continue
                
                # Check rows
                if val in rows[r]:
                    return False
                rows[r].add(val)
                
                # Check cols
                if val in cols[c]:
                    return False
                cols[c].add(val)
                
                # Check box
                if val in subbox[r//3][c//3]:
                    return False
                subbox[r//3][c//3].add(val)

        return True
````

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        cur = ans = ListNode()
        carry = 0

        while l1 or l2 or carry:
            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0
            total = v1 + v2 + carry
            carry = total // 10
            cur.next = ListNode(total % 10)
            cur = cur.next
            
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next
        return ans.next
````

````python
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        # There are several things needed to be considered:
        # 1. If the total gas is less than total cost, we cannot complete the circuit
        # 2. When we reach the bottom, we cannot go down more but go up, so once we find the bottom index, we start from the next index, and we cannot go lower.
        # 3. If total gas >= total cost, there must be a solution, and it is unique. 

        
        # =====================================
        
        # This is O(n^2) which is TLE
        n = len(gas)
        left = 0
        right = left + n - 1

        gas = gas + gas
        cost = cost + cost
        

        while left < n:
            use = 0
            ok = True
            for i in range(left, right + 1):
                use += gas[i] - cost[i]
                if use < 0:
                    ok = False
                    break
            if ok:
                return left
            left += 1
            right += 1

        return -1
````

If there is **no common value**, the function should return `-1`.

Example

    A[0] = 1      B[0] = 4
    A[1] = 3      B[1] = 2
    A[2] = 2      B[2] = 5
    A[3] = 1      B[3] = 3

````python
from collections import Counter

def Solution(A, B):
    # Convert list to set which take O(n) time
    ans = float("inf")
    set_B = set(B)
    for i in A:
        if i in set_B:
            ans = min(ans, i)
            
    return ans

print(Solution([3,2, 5, 9, 2, 3, 2], [3,2,1,4,5,2]))


# If lists are sorted, we can use two pointers to find the minimum common element in O(m+n) time

def Solution_sorted(A, B):
    A = sorted(A)
    B = sorted(B)
    i, j = 0, 0
    ans = float("inf")
    
    while i < len(A) and j < len(B):
        if A[i] == B[j]:
            ans = min(ans, A[i])
            i+=1
            j+=1
            
        elif A[i] > B[j]:
            j+=1
        else:
            i+=1

    return ans if ans != float("inf") else -1

print(Solution_sorted([3, 2, 5, 9, 2, 3, 2], [3,2,1,4,5,2]))
print(Solution_sorted([3, 5], [3]))
print(Solution_sorted([1,2,3], [4,5,6]))
````

#### ***HSBC_3***
Given a string `S` (only `'a'` / `'b'`), return how many distinct substrings of length 3 appear in `S`.  
If `|S| < 3`, return `0`.

### Example 1

S = "bababb"  
→ three-letter substrings: **bab**, **aba**, **bab**, **abb**  
→ distinct = {bab, aba, abb}  
**Output:** 3


S = "bbbaaab"  
→ substrings: **bbb**, **bba**, **baa**, **aaa**, **aab**  
→ distinct = 5  
**Output:** 5


S = "ab"  
→ no length-3 substring  
**Output:** 0

````python
def Solution(S):
    if len(S) < 3:
        return 0
    if len(S) == 3:
        return 1
    
    left = 0
    seen = set()
    
    while left < len(S)-2:
        right = left + 2
        substring = S[left:right+1]
        if substring not in seen:
            seen.add(substring)
            
        left += 1
            
    return len(seen)

print(Solution("bbbaaab"))
````

#### ***HSBC_4***

You are given an integer N.
Create an array A consisting of exactly N integers that satisfies the following conditions:

The sum of all elements in A must be exactly 0.

All elements in A must have unique absolute values.

In other words, if an element x is in A, then neither x nor -x may appear again.

The absolute value of every number in A must not exceed the specified limits (used for scoring).

Scoring (you can ignore if just solving for correctness)

Your solution will be tested under five different limits for the absolute values of elements in A (e.g., ≤ 1 000 000, ≤ N, ≤ 5 × N, etc.).
Performance is not the focus — correctness is.

Example:

Input:
N = 4

Possible Outputs:
[0, -4, 1, 3] or [-1, 0, -3, 4] or [1, -1, 2, -2] or [5, 6, 0, -11] Only one valid list is fine

### Proof

It is very important to know if all value of N has at least one valid output.

When N == 2, two element x and y, x+y = 0, y = -x which against unique absolute values.

If N >= 3, we let S = sum(1, 2, 3, ... N-1)

so S = ((1 + (N-1)) * (N-1)) // 2 and S is strictly larger than N-1, so it does not collide with {1, N-1}

sum(1, 2, 3, ... N-1) + (-S) == 0


````python
def Solution(N):
    if N == 1:
        return [0]
    
    if N <= 2:
        return []
    
    S = (1 + N-1) * (N-1) // 2
    ans = list(range(1, N)) + [-S]
    return ans

print(Solution(5))
print(Solution(20))
````

#### ***HSBC_6***

You are given two strings `S` and `T`, both of length `N`.  
Each string consists only of digits (`'0'–'9'`).

In a single move, you can choose an index `K` in the range `[0, N−2]` and **increase both digits `S[K]` and `S[K+1]` by 1**.  
If a digit is `9` and you increase it, it becomes `0` (mod 10).

Example:

Input: S = "13471", T = "59604"
Output: 9

S = "557", T = "445"

| Step | Operation | Result |
|:----:|:-----------|:--------|
| 0 | Initial | 557 |
| +1 | increase(0) | 667 |
| +2 | increase(0) | 778 |
| +3 | increase(0) | 889 |
| +4 | increase(0) | 990 |
| +5 | increase(0) | 001 |
| +6 | increase(0) | 112 |
| +7 | increase(0) | 223 |
| +8 | increase(0) | 334 |
| +9 | increase(0) | 445 |

Tips: 

- Every move must **act on index `i` and `i+1` together**.  
- The operation is **cyclic modulo 10**.  
- You need to find the **minimum number of moves** to transform `S` into `T`.

````python
def solution(S, T):
    S = [int(x) for x in S]
    T = [int(x) for x in T]
    n = len(S)
    ans = 0

    for i in range(n - 1):
        diff = (T[i] - S[i]) % 10
        S[i] = (S[i] + diff) % 10
        S[i + 1] = (S[i + 1] + diff) % 10
        ans += diff

    if S[-1] != T[-1]:
        return -1

    return ans
````

#### ***HSBC_7***

You are given a string `S`, which consists entirely of decimal digits (`0–9`).  
Using digits from `S`, create a **palindromic number** with the **largest possible decimal value**.  
You should use at least **one** digit and you **can reorder** the digits.

A palindromic number remains the same when its digits are reversed; for instance, `"7"`, `"44"` or `"83238"`.  
Any palindromic number you create **must not have leading zeros**, such as `"0990"` or `"010"`.

For example, decimal palindromic numbers that can be created from `"8199"` are:  
`"1"`, `"8"`, `"9"`, `"9"`, `"919"` and `"989"`.  
Among them, `"989"` has the largest value.

````python
def solution(S):
    cnt = {}
    for ch in S:
        cnt[ch] = cnt.get(ch, 0) + 1

    left_parts = []
    for i in range(9, -1, -1):
        d = str(i)
        pairs = cnt.get(d, 0) // 2
        if pairs > 0:
            left_parts.append(d * pairs)
            cnt[d] = cnt.get(d, 0) - pairs * 2

    centre = ''
    for i in range(9, -1, -1):
        d = str(i)
        if cnt.get(d, 0) % 2 == 1:
            centre = d
            break

    left_str = ''.join(left_parts)
    ans = left_str + centre + left_str[::-1]

    j = 0
    while j < len(ans) and ans[j] == '0':
        j += 1
        
    if j == len(ans):
        
        return '0'
    return ans[j:]
````
